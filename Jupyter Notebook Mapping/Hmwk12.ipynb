# Hmwk12 - ISS Position Project

This is a project which utilizes Cartopy to show the current position of the ISS every time the program is run.

```python
# Here I import what modules I will need, notably json, cartopy, and matplotlib
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import json
from urllib.request import urlopen

# Here I create the map and the gridlines on said map
plt.figure(figsize = (12, 8))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.stock_img()
ax.gridlines(draw_labels=True)

# Here I get the raw data from a url trackin the ISS and then upload it via json.
# Then a variable named "f" is made equal to the array from the website
url = "http://api.open-notify.org/iss-now.json"
response = urlopen(url)
f = json.loads(response.read())

# Here I made a variable so I could extract the longitude and latitude separately
longitudeAndLatitude = f["iss_position"]

# Here I extract the longitude and latitude separately from the variable made in the last step.
# For ax.plot() to work you have to change the values to floats
latitude = float(longitudeAndLatitude["latitude"])
longitude = float(longitudeAndLatitude["longitude"])

# Printing where the position is for the user
print ("ISS latitude position is: ", latitude)
print ("ISS longitude position is: ", longitude)

# The actual point for the ISS!
ax.plot(longitude, latitude,
         color='red', linewidth=2, marker='o')

plt.show()
```
